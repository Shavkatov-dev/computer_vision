```python
import matplotlib.pyplot as plt #pyplot sub-modul
import numpy as np
```

# **Basic Line Plot**

```python
x = np.array([1, 2, 3, 4,])
y = np.array([10, 20, 50, 40])

plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python

x= np.array([1,2 ,3, 4])
y = np.array([3, 8, 1, 10])

plt.plot(y, linestyle='dotted', color='green', marker='x')
plt.title("Styles Line Example")
plt.show()
```

# **Multiple Lines with One Graph with Numpy and MatPlotLib**

```python
x= np.array([1, 2 ,3, 4])
y1 = np.array([2, 4, 6, 8])
y2 = np.array([1, 3, 5, 7])

plt.plot(x, y1, label='Line 1', color='red', marker='s')
plt.plot(x, y2, label='Line 2', color='green', linestyle='--', marker='o')
plt.title("Multiple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

# **Bar Chart with Numpy and MatPlotLib**

```python
x = np.array(['Abba', 'Bilol', 'Clark', 'David'])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color='orange')
plt.title("Basic Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

# **Scatter Plot with Numpy and MatPlotLib**

```python
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4])
y = np.array([3, 13, 14, 5, 20, 23, 4, 12,1])

plt.scatter(x, y, color='purple', s=100)
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# **Histogram with Numpy and MatPlotLib**

```python
data = np.array([3,4,5,6,7,8,9,1])

plt.hist(data, bins=5, color='skyblue',edgecolor="black") #
plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel("Count")
plt.show()
```

# **Subplots(Multiple Graphs Together)**

```python
x= np.array([1, 2 ,3, 4])
y1 = np.array([2, 4, 6, 8])
y2 = np.array([1, 3, 5, 7])

plt.figure(figsize=(10, 5)) # width, height

plt.subplot(1, 2, 1)
plt.plot(x, y1, color='red', marker='s')
plt.title("Line Plot 1")

plt.subplot(1, 2, 2)
plt.bar(x, y2, color='blue')
plt.title("Line Plot 2")

plt.tight_layout()
plt.show()
```

```python
x = np.array([5, 3, 8, 4])
y = np.array([3, 4, 6, 8])

plt.plot(x, y, color='green')
plt.title("Save Figure Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)

plt.savefig('my_plot.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# make a stream function:
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
# make U and V out of the streamfunction:
V = np.diff(Z[1:, :], axis=1)
U = -np.diff(Z[:, 1:], axis=0)

# plot:
fig, ax = plt.subplots()

ax.streamplot(X[1:, 1:], Y[1:, 1:], U, V)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# Make data
n = 100
xs = np.linspace(0, 1, n)
ys = np.sin(xs * 6 * np.pi)
zs = np.cos(xs * 6 * np.pi)

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot(xs, ys, zs)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# Make data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap="Blues")

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```
