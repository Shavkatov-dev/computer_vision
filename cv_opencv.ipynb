```python
import cv2
```

```python
from google.colab.patches import cv2_imshow
image = cv2.imread("/content/cat-image.jpg")
```

```python
cv2_imshow(image)
```

```python
print("shape:", image.shape)
#height, weight, color channels(RGB) => (BGR)

print("HEIGHT:", image.shape[0])
print("WIDTH:", image.shape[1])
print("CHANNELS:", image.shape[2])
```

```python
(b, g, r) = image[100, 50] #
print("Blue:", b, "Green:", g, "Red:" , r)
```

```python
GRAY = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(GRAY)
```

```python
resized = cv2.resize(GRAY,(300, 300))
cv2_imshow(resized)
```

```python
cropped = GRAY[100:300, 150:350]
cv2_imshow(cropped)
```

```python
rotated = cv2.rotate(GRAY, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(rotated)
```

```python
flipped = cv2.flip(GRAY, 1)
cv2_imshow(flipped)
```

```python
cv2.imwrite("flipped-cat.jpg", flipped)
```
